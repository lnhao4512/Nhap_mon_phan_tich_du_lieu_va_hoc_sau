Vấn đề 7: “Missing values”, vấn đề này xảy ra tại các cột “Age”, “Weight” và “Heart
Rate”. Thiếu dữ liệu (dữ liệu không đầy đủ) là vấn đề xảy ra nhiều trong các nguồn dữ
liệu do nhiều nguyên nhân chủ quan lẫn khách quan.Có một vài giải pháp để xử lý vấn đề
này, chủ yếu dựa trên kinh nghiệm và nghiệp vụ về tập dữ liệu đó. Một số giải pháp đưa
đề xuất từ chuyên gia như sau:
a. Deletion: Remove records with missing values
b. Dummy substitution: Replace missing values with a dummy but valid value:
e.g.: 0 for numerical values.
c. Mean substitution: Replace the missing values with the mean.
d. Frequent substitution: Replace the missing values with the most frequent item.
e. Improve the data collector: Your business folk will talk to the clients and inform
them about why it is worth fixing the problem with the data collector.
……
Yêu cầu:
- Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight
- Yêu cầu xử lý dữ liệu thiếu như sau: Nếu dòng nào có Age hoặc Weight có dữ liệu thì
phần Age hoặc Weight được tính như bên dưới, nếu thiếu cả 2 thông tin thì xóa dòng
o Age: Giá trị thay thế là mean của các giá trị trong cột Age

```python
# Vấn đề 7: Missing values

# 0. Chuyển cột Weight sang dạng số (loại bỏ 'kgs' hoặc 'kg')
df.loc[:, 'Weight'] = df['Weight'].astype(str).str.replace('kgs', '').str.replace('kg', '')
df.loc[:, 'Weight'] = pd.to_numeric(df.loc[:, 'Weight'], errors='coerce')

# 1. Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight
print("=== Thống kê dữ liệu thiếu ===")
print(f"Số giá trị thiếu trong cột Age: {df['Age'].isna().sum()}")
print(f"Số giá trị thiếu trong cột Weight: {df['Weight'].isna().sum()}")
print(f"\nTổng số dòng: {len(df)}")
print(f"Số dòng thiếu cả Age và Weight: {(df['Age'].isna() & df['Weight'].isna()).sum()}")

# 2. Xử lý dữ liệu thiếu
# Bước 1: Xóa các dòng thiếu cả Age và Weight
df = df[~(df['Age'].isna() & df['Weight'].isna())]

# Bước 2: Điền giá trị mean cho cột Age nếu thiếu
age_mean = df['Age'].mean()
df.loc[:, 'Age'] = df['Age'].fillna(age_mean)

# Bước 3: Điền giá trị mean cho cột Weight nếu thiếu
weight_mean = df['Weight'].mean()
df.loc[:, 'Weight'] = df['Weight'].fillna(weight_mean)

print("\n=== Sau khi xử lý ===")
print(f"Số giá trị thiếu trong cột Age: {df['Age'].isna().sum()}")
print(f"Số giá trị thiếu trong cột Weight: {df['Weight'].isna().sum()}")
print(f"Tổng số dòng còn lại: {len(df)}")
print(df)
```

Vấn đề 8: “một cột chứa quá nhiều thông tin cần được phân rã”, như trong bài toán này ta
thấy header “m0006” chứa các nội dung bao gồm: m → male, 1218 ~ 12-18 (mm-dd).
Còn giá trị thì là kết quả huyết áp.

```python
df = pd.melt(
    df,
    id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'],
    value_name='PulseRate',
    var_name='sex_and_time'
).sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

# Extract Sex, hour lower bound and hour upper bound
tmp_df = df['sex_and_time'].str.extract(r'(\D+)(\d{2})(\d{2})', expand=True)

# Rename columns
tmp_df.columns = ['sex', 'hours_lower', 'hours_upper']

# Create Time column based on hours_lower and hours_upper columns
tmp_df['Time'] = tmp_df['hours_lower'] + '-' + tmp_df['hours_upper']

# Merge
df = pd.concat([df, tmp_df], axis=1)

# Drop unnecessary columns and rows
df = df.drop(['sex_and_time', 'hours_lower', 'hours_upper'], axis=1)
df = df.dropna()

df.to_csv('outputcleanup.csv', index=False)
print(df)
```

11. Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp. Dữ liệu bị thiếu thì hãy xử lý bằng
phương pháp sau
• Thay thế bằng giá trị trung bình liền trước và liền sau của người đó. Nếu không
được thì dùng 2)
• Thay thế bằng giá trị trung bình 2 giá liền trước của người đó. Nếu không được
thì dùng 3)
• Thay thế bằng giá trị trung bình 2 giá liền sau của người đó. Nếu không được thì
dùng 4)
• Trung bình của các giá trị huyết áp của người đó. Nếu không được thì dùng 5).
• Trung bình của các giá trị huyết áp của nhóm giới tính. Nếu không được thì dùng
6)
• Trung bình của các giá trị dữ liệu. Nếu không được thì thay bằng mức ổn định
trong y học.

```python
# Vấn đề 11: Xử lý dữ liệu thiếu trên biến PulseRate

# Chuyển PulseRate sang dạng số
df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')

# Khảo sát tỉ lệ dữ liệu thiếu
total_rows = len(df)
missing_count = df['PulseRate'].isna().sum()
missing_rate = (missing_count / total_rows) * 100

print("=== Khảo sát dữ liệu thiếu trên PulseRate ===")
print(f"Tổng số dòng: {total_rows}")
print(f"Số giá trị thiếu: {missing_count}")
print(f"Tỉ lệ thiếu: {missing_rate:.2f}%")

# Giá trị huyết áp ổn định trong y học (nhịp tim bình thường: 60-100 bpm)
NORMAL_PULSE = 75

# Hàm xử lý missing values theo thứ tự ưu tiên
def fill_pulse_rate(group):
    group = group.sort_values('Time').reset_index(drop=True)
    pulse = group['PulseRate'].copy()

    for i in range(len(pulse)):
        if pd.isna(pulse.iloc[i]):
            # 1. Trung bình liền trước và liền sau
            if i > 0 and i < len(pulse) - 1:
                prev_val = pulse.iloc[i-1]
                next_val = pulse.iloc[i+1]
                if pd.notna(prev_val) and pd.notna(next_val):
                    pulse.iloc[i] = (prev_val + next_val) / 2
                    continue

            # 2. Trung bình 2 giá trị liền trước
            if i >= 2:
                prev1 = pulse.iloc[i-1]
                prev2 = pulse.iloc[i-2]
                if pd.notna(prev1) and pd.notna(prev2):
                    pulse.iloc[i] = (prev1 + prev2) / 2
                    continue

            # 3. Trung bình 2 giá trị liền sau
            if i < len(pulse) - 2:
                next1 = pulse.iloc[i+1]
                next2 = pulse.iloc[i+2]
                if pd.notna(next1) and pd.notna(next2):
                    pulse.iloc[i] = (next1 + next2) / 2
                    continue

            # 4. Trung bình các giá trị huyết áp của người đó
            person_mean = pulse.dropna().mean()
            if pd.notna(person_mean):
                pulse.iloc[i] = person_mean
                continue

            # Đánh dấu cần xử lý theo giới tính (sẽ xử lý sau)
            pulse.iloc[i] = np.nan

    group['PulseRate'] = pulse.values
    return group

# Áp dụng hàm xử lý theo từng người (Firstname, Lastname)
df = df.groupby(['Firstname', 'Lastname'], group_keys=False).apply(fill_pulse_rate, include_groups=False)

# Thêm lại các cột Firstname, Lastname nếu bị mất
if 'Firstname' not in df.columns:
    df = df.reset_index()

# 5. Trung bình theo nhóm giới tính
sex_means = df.groupby('sex')['PulseRate'].transform('mean')
df['PulseRate'] = df['PulseRate'].fillna(sex_means)

# 6. Trung bình toàn bộ dữ liệu hoặc mức ổn định y học
overall_mean = df['PulseRate'].mean()
if pd.notna(overall_mean):
    df['PulseRate'] = df['PulseRate'].fillna(overall_mean)
else:
    df['PulseRate'] = df['PulseRate'].fillna(NORMAL_PULSE)

print("\n=== Sau khi xử lý ===")
print(f"Số giá trị thiếu còn lại: {df['PulseRate'].isna().sum()}")
print(df)
```

Lưu ý: Ngoài ra còn rất nhiều vấn đề về mặt xử lý dữ liệu dựa trên nhiều khía cạnh khác
nhau tùy vào sự am hiểu về dữ liệu của các chuyên gia như:

• Correcting character encodings (a problem you hit when you scrape data off the

```python
# Vấn đề 12: Rút gọn & sắp xếp lại dữ liệu

print("Cột hiện có:", df.columns.tolist())

# Reset index nếu thiếu cột quan trọng
if 'Firstname' not in df.columns or 'Lastname' not in df.columns:
    df = df.reset_index()
    print("Đã reset index để khôi phục Firstname/Lastname")

# Thứ tự cột mong muốn (chỉ giữ cột tồn tại)
desired = ['Id', 'Firstname', 'Lastname', 'Age', 'Weight', 'sex', 'Time', 'PulseRate']
df = df[[c for c in desired if c in df.columns]]

# Sắp xếp theo các cột có sẵn
sort_by = [c for c in ['Id', 'Firstname', 'Lastname', 'sex', 'Time'] if c in df.columns]
if sort_by:
    df = df.sort_values(sort_by)

# Reset index sạch sẽ
df = df.reset_index(drop=True)

# Làm tròn số thập phân (nếu có cột)
for col in ['Age', 'Weight', 'PulseRate']:
    if col in df.columns:
        df[col] = df[col].round(1)

# In thông tin tóm tắt
print("\n=== Kết quả sau khi xử lý ===")
print(f"- Số dòng: {len(df)}")
print(f"- Số cột: {len(df.columns)}")
print(f"- Cột: {df.columns.tolist()}")
print("\nKiểu dữ liệu:\n", df.dtypes)
print("\nDữ liệu mẫu (10 dòng đầu):\n", df.head(10))

# Lưu file
df.to_csv('patient_heart_rate_clean.csv', index=False, encoding='utf-8')
print(f"\nĐã lưu thành công: patient_heart_rate_clean.csv")
print(f"Tổng số bản ghi: {len(df)}")
```

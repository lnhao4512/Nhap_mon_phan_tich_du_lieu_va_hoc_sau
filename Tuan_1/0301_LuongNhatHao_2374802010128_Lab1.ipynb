```python
import pandas as pd #Lương Nhật Hào - 2374802010128
import numpy as np
```

#Tải dữ liệu xét tuyển đại học

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT") #tải dữ liệu lên chương trình
df.head(10) #in ra màn hình 10 dòng đầu tiên
```

```python
df.tail(10) #in ra màn hình 10 dòng cuối cùng
```

```python
#Câu 4:
#Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị 0.
df['DT'].fillna(0, inplace=True)
df
```

```python
#Câu 5:
#Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean
df["T1"].fillna(df["T1"].mean(), inplace=True)
df
```

```python
#Câu 6:
#Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
score_columns = ['L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1', 'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2', 'L3', 'H3', 'S3', 'V3', 'X3', 'D3', 'N3', 'L4', 'H4', 'S4', 'V4', 'X4', 'D4', 'N4', 'L5', 'H5', 'S5', 'V5', 'X5', 'D5', 'N5', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6', 'DH1', 'DH2', 'DH3']
for col in score_columns:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
#Câu 7:
#Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
#• Công thức tính: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T3']*2 + df['L3'] + df['H3'] + df['S3'] + df['V3']*2 + df['X3'] + df['D3'] + df['N3']) / 10
df.head()
```

```python
#Câu 8
#Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:
#• Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#• Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#• Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#• Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#• Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)
def xep_loai(trungbinhmon):
    if trungbinhmon < 5.0:
        return 'Y'
    elif 5.0 <= trungbinhmon < 6.5:
        return 'TB'
    elif 6.5 <= trungbinhmon < 8.0:
        return 'K'
    elif 8.0 <= trungbinhmon < 9.0:
        return 'G'
    else:
        return 'XS'
df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
df.head()
```

```python
#Câu 9:
#Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ.
#Sử dụng phương pháp Min-Max Normalization
df['US_TBM1'] = df['TBM1'] * 0.4
df['US_TBM2'] = df['TBM2'] * 0.4
df['US_TBM3'] = df['TBM3'] * 0.4
df.head()
```

```python
#Câu 10:
#. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
#• Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
#• Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
#• Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
def tinhtoan_kqxt(row):
    if row['KT'] in ['A', 'A1']:
        score = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        score = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        score = (row['DH1'] + row['DH2'] + row['DH3']) / 3

    if score >= 5.0:
        return 1  # Đậu
    else:
        return 0  # Rớt
df['KQXT'] = df.apply(tinhtoan_kqxt, axis=1)
df.head()
```

```python
#Câu 11:
#Lưu trữ dữ liệu xuống ổ đĩa thành file processed_dulieuxettuyendaihoc.csv
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("DataFrame đã được lưu vào 'processed_dulieuxettuyendaihoc.csv'")
```
